step1- Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

step2 creating data frame

```python
dataframe = pd.read_csv("ElectricCarData_Clean.csv")
print(dataframe)
```

# Performance KPIs

```python
sns.countplot(x=dataframe ['BodyStyle'])
plt.xlabel("Based on the Body Type")
```

# conclusion - Majority of the people choose the suv body type

```python
dataframe.head()
```

```python
group_data = dataframe.groupby('BodyStyle')['TopSpeed_KmH'].sum()
result = pd.DataFrame({'TopSpeed_KmH':group_data})
plt.plot(result, c="red",marker="*")
plt.xlabel("Type of Body",c="green",size=20)
plt.ylabel("Top speed in km/Hrs",c="green",size=20)
```

# conclusion SUV has maximum speed per hours


```python
dataframe.head()
```

```python
plt.hist(dataframe['Range_Km'],bins=5)
plt.title("Range in Kilometers")
plt.show()
```

# conclusion majority cars give 150 to 400 kms

```python
couple_data = dataframe["Seats"]
sns.countplot(x=couple_data)
```

# conclusion mojority of people choose 5 Seater

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='BodyStyle',columns='PowerTrain',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True,cmap="YlGnBu",fmt='d')
plt.title("Heatmap")
plt.xlabel("Brand")
plt.ylabel("BodyStyle")
plt.show()
```

```python
pivot_table = dataframe.pivot_table(index='BodyStyle',columns='Brand',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True,cmap="YlGnBu",fmt='d')
plt.title("Heatmap")
plt.xlabel("Brand")
plt.ylabel("BodyStyle")
plt.show()
```
